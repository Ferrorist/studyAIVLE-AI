## <b>Q2. html 에서 \<script>...\</script> 태그 길이 계산
- BeautifulSoup으로 html소스를 python 객체로 변환
- 함수로 구현하기
- float으로 return 받기

### <span style="color:darkred">[문제2] Beatuifulsoup 라이브러리를 활용 HTML code에서 \<script> 태그 길이를 계산하는 함수를 완성하고 결과를 확인하세요.<span>

```python
soup = BeautifulSoup(df['html_code'][0], 'html.parser')
```

```python
soup.head
```

```python
print(len(str(soup.head.script.getText())))
print(len(str(soup.body.script.getText())))
```

```python
# Feature(특징) 데이터를 추출하는 함수를 작성합니다.
def head_script_length(soup): # head의 script 태그 길이
    try:
        return len(str(soup.head.script.getText()))
    except:
        return 0.0
    
def body_script_length(soup): # body의 script 태그 길이
    try:
        return len(str(soup.body.script.getText()))
    except:
        return 0.0
    
def script_length(soup):
    return head_script_length(soup) + body_script_length(soup)
```

```python
# 데이터 프레임의 html_code 컬럼에서 Feature(특징) 데이터를 추출합니다.
script_len = []

for index, row in df.iterrows():
    soup = BeautifulSoup(row.html_code, 'html.parser')
    script_len.append(script_length(soup))
```

```python
# 추출한 Feature(특징) 데이터를 확인합니다.
df['script_length'] = script_len
df.head()
```

### <span style="color:darkred">[문제5] Beatuifulsoup 라이브러리를 활용 HTML code에서 \<script> 태그에서 src, href 속성을 가진 태그수를 계산하는 함수를 완성하고 결과를 확인하세요. <span>


```python
print(str(soup.body).count('src'))
print(str(soup.body).count('href'))
```

```python
# Feature(특징) 데이터를 추출하는 함수를 작성합니다.
def count_src_href(soup):
    temp = str(soup.body)
    return temp.count('src') + temp.count('href')
```

```python
# 데이터 프레임의 html_code 컬럼에서 Feature(특징) 데이터를 추출합니다
src_href = []

for index, row in df.iterrows():
    soup = BeautifulSoup(row.html_code, 'html.parser')
    src_href.append(count_src_href(soup))
```

```python
# 추출한 Feature(특징) 데이터를 확인합니다.
df['count_src_href'] = src_href
```

### <span style="color:darkred">[문제6] Beatuifulsoup 라이브러리를 활용 HTML code에서 추가로 만들수 있는 Feature를 찾아보고 결과를 확인하세요. <span>


```python
soup = BeautifulSoup(df['html_code'][0], 'html.parser')
```

```python
divs = soup.body.find_all('div')
count_div(soup)
```

```python
# Feature(특징) 데이터를 추출하는 함수를 작성합니다.
def count_div(soup):
    count = 0
    try:
        divs = soup.body.find_all('div')
        for div in divs:
            count += 1
        return count
    except:
        return 0.0
    
def count_div1(soup):
    count = 0
    divs = soup.body.find_all('div')
    for div in divs:
        count += 1
    return count
```

```python
# 데이터 프레임의 html_code 컬럼에서 Feature(특징) 데이터를 추출합니다
div_count = []

for index, row in df.iterrows():
    soup = BeautifulSoup(row.html_code, 'html.parser')
    div_count.append(count_div(soup))
```

```python
# 추출한 Feature(특징) 데이터를 확인합니다.
df['count_div'] = div_count
df
```
